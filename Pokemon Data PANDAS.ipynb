```python
import pandas as pd
```

```python
poke = pd.read_csv('poke data.csv')

print(poke.head(2))
```

```python
#Read Headers
print(poke.columns)
print(poke.head(10))
#Read Each Column
print(poke[['Generation', 'Attack']])

#Read Each Row
print(poke.iloc[9:11])

#Read Specific Location
print(poke.iloc[1,3])
```

```python
# Sorting and Describing Data

poke.describe()

poke.sort_values('Name') #Alphabetic

poke.sort_values('Name', ascending = False) #Non-alphabetic

poke.sort_values(['Type 2', 'Attack'], ascending = [1,0])

```

```python
#Making Changes to the Data
poke['Total'] = poke['HP'] + poke['Attack'] + poke['Defense'] + poke['Sp. Atk'] + poke['Sp. Def'] + poke['Speed']

poke['Total'] = poke.iloc[:, 4:10].sum(axis = 1)

poke.head(5)
```

```python
# Saving/ Exporting our Data

poke.to_csv('PokeUpdated.csv', index = False)
```

```python
#Filtering Data

poke.loc[(poke['Type 1'] == 'Fire') & (poke['Attack'] >= 100)]

poke.loc[poke['Name'].str.contains('Mega')]

# Setting new DF and then pulling data from new DF

pokefire = poke.loc[(poke['Type 1'] == 'Fire') & (poke['Attack'] >= 100)]

pokefire.sort_values('Attack', ascending = False)

pokefire.head(3)

pokefire.loc[pokefire['Attack'] >100]

pokefire.iloc[2,-4]

```

```python
# Changing data with Loc function

poke.loc[poke['Total'] > 600, ['Legendary']] = "Super Legendary"
poke
```

```python
# Group by

poke.groupby('Generation').sum().sort_values('Attack', ascending = False )
```
